```python
# грузим библиотеки
from tensorflow import keras
from google.colab import files
from io import BytesIO
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# грузим УЖЕ ОБУЧЕННУЮ модельку нейросети VGG16 с подобранными весами (одни веса весят over 500Mb!)
model = keras.applications.VGG16()
```

```python
# грузим изображение 224х224, которое принимает на вход нейросеть (можно изменить размер средствами Win10)
uploaded = files.upload()
```

```python
# открываем загруженное изображение
img = Image.open(BytesIO(uploaded['ex224.jpg']))
```

```python
# отрисовываем изображение
plt.imshow( img )
```

```python
# изменяем изображение, сначала создав из него массив numpy, затем этот массив предпроццессируется спец. методом,
# меняя изображение с RGB на BGR и центрирует все значения массива вокруг нуля
img = np.array(img)
x = keras.applications.vgg16.preprocess_input(img)
```

```python
# меняем форму тензора с 3-х мерного на 4-х мерный
print(x.shape)
x = np.expand_dims(x, axis=0)
```

```python
# заводим в модель обработанные данные и выводим результат в виде числа-классификатора
res = model.predict( x )
print(np.argmax(res))
```
