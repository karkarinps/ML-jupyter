```python
import pandas as pd
df = pd.read_csv('C://Users/Uldum/Downloads/out.csv')
df.info()
```

```python
df.duplicated().unique()
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
for column in df.columns:
    try:
        plt.figure()
        df.boxplot([column])
    except:
        print("Smth wrong")
```

```python
df_save=df
for column in df.columns:
  limiter = df[column].quantile(0.99)
  df = df[df[column]<=limiter]
    
df.info()
```

```python
df=df_save
df_dum = pd.get_dummies(df)
df_dum.info()
```

```python
df_dum = df_dum.drop(['Unnamed: 0', 'SPP'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler, normalize

scaler = StandardScaler()
scaler.fit(df_dum)
df_scale = pd.DataFrame(scaler.transform(df_dum), columns = df_dum.columns)
df_scale.head(10)
```

```python
df_norm = pd.DataFrame(normalize(df_scale), columns = df_scale.columns)
df_norm.head(10)
```

```python
import seaborn as sns

cor = df_norm.corr()
sns.heatmap(cor, square = True)
```

```python
df_norm['Promo Sale'].value_counts()
```

```python
df_norm = df_norm.drop(['Promo Sale'], axis=1)
cor = df_norm.corr()
sns.heatmap(cor, square = True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X = df_norm.drop(['SPP Price'], axis=1) 
y = df_save.SPP        
y 
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_reg = Ridge(alpha = 1.0)
ridge_reg.fit(X, y)
```

```python
imp = pd.DataFrame(ridge_reg.coef_, index=X.columns, columns=['importance'])
imp.sort_values('importance').plot(kind='barh', figsize=(12, 8))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg1 = LogisticRegression(solver='sag', random_state=12)
logreg1.fit(X, y) 
```

```python
from sklearn.inspection import permutation_importance
```

```python
model_fi = permutation_importance(logreg1, X, y)
model_fi['importances_mean']
```

```python
imp = pd.DataFrame(model_fi['importances_mean'], index=X.columns, columns=['importance'])
imp.sort_values('importance').plot(kind='barh', figsize=(12, 8))
```

```python
feat_list = ['Base price', 'Basic Sale', 'Median price', 'Min price', 'Balance FBS', 'FBS', 'Median price', 'Revenue potential', 'Days with sales', 'sum_stocks']
```

```python
X = df_norm.drop(['SPP Price'], axis=1)[feat_list] 
y = df_save.SPP        
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=13)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

extc = ExtraTreesClassifier(n_estimators=80, max_features= 128, criterion= 'entropy', min_samples_split= 4,
                            max_depth= 14, min_samples_leaf= 3)      
extc.fit(X_train,y_train)
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
```

```python
f1_score(y_test, extc.predict(X_test), average='macro')
```

```python
roc_auc_score(y, extc.predict_proba(X), average='macro', multi_class='ovr')
```

```python
imp = pd.DataFrame(extc.feature_importances_, index=X_train.columns, columns=['importance'])
imp.sort_values('importance').plot(kind='barh', figsize=(12, 8))
```
